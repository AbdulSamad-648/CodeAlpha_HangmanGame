## Section 1: Import Required Libraries

We import the `random` module to randomly select a word from our predefined list.

```python
import random

print("✅ Libraries imported successfully!")
```

## Section 2: Define the Word List

Create a list of 5 predefined words that will be randomly selected for each game.

```python
# Predefined word list for Hangman game
WORD_LIST = ['python', 'hangman', 'computer', 'programming', 'internship']

print(f"📚 Word list created with {len(WORD_LIST)} words:")
for i, word in enumerate(WORD_LIST, 1):
    print(f"   {i}. {word}")
```

## Section 3: Initialize Game Variables

Set up the initial game state including the selected word, guessed letters, and game counters.

```python
def initialize_game():
    """
    Initialize game variables for a new game.
    Returns: (word, word_length, guessed_letters, guessed_word, tries)
    """
    word = random.choice(WORD_LIST).upper()
    word_length = len(word)
    guessed_letters = []
    guessed_word = ['_' for _ in range(word_length)]
    tries = 6
    
    return word, word_length, guessed_letters, guessed_word, tries

# Test the initialization
word, word_length, guessed_letters, guessed_word, tries = initialize_game()
print(f"✅ Game initialized!")
print(f"   Word length: {word_length}")
print(f"   Tries remaining: {tries}")
print(f"   Word display: {' '.join(guessed_word)}")
```

## Section 4: Display Hangman Stages

Create ASCII art to visualize the hangman stages based on remaining tries.

```python
def display_hangman(tries):
    """
    Display the hangman figure based on remaining tries.
    """
    stages = [  # Final state
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |     / \\
           -
        """,
        # Head, torso, both arms, one leg
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |     / 
           -
        """,
        # Head, torso, both arms
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |      
           -
        """,
        # Head, torso, one arm
        """
           --------
           |      |
           |      O
           |     \\|
           |      |
           |     
           -
        """,
        # Head and torso
        """
           --------
           |      |
           |      O
           |      |
           |      |
           |     
           -
        """,
        # Head only
        """
           --------
           |      |
           |      O
           |    
           |      
           |     
           -
        """,
        # Initial state
        """
           --------
           |      |
           |      
           |    
           |      
           |     
           -
        """
    ]
    return stages[tries]

# Test the hangman display
print("Hangman display with 3 tries remaining:")
print(display_hangman(3))
```

## Section 5: Display Game State

Create a function to display the current game state with all relevant information.

```python
def display_game_state(guessed_word, guessed_letters, tries):
    """
    Display the current game state.
    """
    print("\n" + "="*50)
    print(f"Word to guess: {' '.join(guessed_word)}")
    print(f"Guessed letters: {', '.join(sorted(guessed_letters)) if guessed_letters else 'None'}")
    print(f"Remaining tries: {tries}")
    print("="*50 + "\n")

# Test the display function
test_guessed_word = ['H', '_', '_', 'G', 'M', '_', 'N']
test_guessed_letters = ['H', 'G', 'M', 'N', 'X', 'Z']
test_tries = 3

display_game_state(test_guessed_word, test_guessed_letters, test_tries)
```

## Section 6: Get and Validate Player Input

Implement input validation to ensure the player enters a valid letter.

```python
def get_valid_guess(guessed_letters):
    """
    Get and validate player input.
    Returns: A valid letter guess
    """
    while True:
        guess = input("Guess a letter: ").upper()
        
        # Validate single character
        if len(guess) != 1:
            print("❌ Please enter exactly one character!")
            continue
        
        # Validate alphabetic
        if not guess.isalpha():
            print("❌ Please enter a letter (A-Z)!")
            continue
        
        # Check if already guessed
        if guess in guessed_letters:
            print(f"❌ You already guessed '{guess}'. Try a different letter.")
            continue
        
        return guess

print("✅ Input validation function ready!")
```

## Section 7: Update Game Logic

Process player guesses and update the game state accordingly.

```python
def process_guess(guess, word, guessed_word, guessed_letters, tries):
    """
    Process the player's guess and update the game state.
    Returns: (updated_guessed_word, updated_tries, is_correct)
    """
    guessed_letters.append(guess)
    
    if guess in word:
        print(f"✅ Correct! '{guess}' is in the word!")
        
        # Update guessed_word with correct guesses
        for i, letter in enumerate(word):
            if letter == guess:
                guessed_word[i] = guess
        
        return guessed_word, tries, True
    else:
        print(f"❌ Wrong! '{guess}' is not in the word.")
        tries -= 1
        return guessed_word, tries, False

print("✅ Game logic function ready!")
```

## Section 8: Main Game Loop

Implement the complete game loop and play a full game!

```python
def play_hangman_game():
    """
    Main function to play a complete game of Hangman.
    """
    # Initialize game
    word, word_length, guessed_letters, guessed_word, tries = initialize_game()
    guessed = False
    
    print("\n" + "="*50)
    print("WELCOME TO HANGMAN GAME!")
    print("="*50)
    print(f"The word has {word_length} letters.")
    print("You have 6 incorrect guesses allowed.")
    print("="*50)
    
    # Game loop
    while not guessed and tries > 0:
        print(display_hangman(tries))
        display_game_state(guessed_word, guessed_letters, tries)
        
        # Get player guess
        guess = get_valid_guess(guessed_letters)
        
        # Process guess
        guessed_word, tries, _ = process_guess(guess, word, guessed_word, guessed_letters, tries)
        
        # Check if player won
        if '_' not in guessed_word:
            guessed = True
    
    # Display final result
    print(display_hangman(tries))
    
    if guessed:
        print("\n" + "="*50)
        print("🎉 CONGRATULATIONS! YOU WON! 🎉")
        print(f"The word was: {word}")
        print(f"Total guesses: {len(guessed_letters)}")
        print(f"Incorrect guesses: {6 - tries}")
        print("="*50)
    else:
        print("\n" + "="*50)
        print("💀 GAME OVER! YOU LOST! 💀")
        print(f"The correct word was: {word}")
        print("="*50)

# Play the game!
play_hangman_game()
```

## Section 9: Play Multiple Rounds

Allow the player to play multiple games in succession.

```python
def main():
    """
    Main function allowing players to play multiple rounds.
    """
    while True:
        play_hangman_game()
        
        # Ask to play again
        play_again = input("\nDo you want to play again? (yes/no): ").lower()
        if play_again not in ['yes', 'y']:
            print("\n" + "="*50)
            print("Thanks for playing! Goodbye! 👋")
            print("="*50 + "\n")
            break

# Uncomment to run the game:
# main()

print("✅ Hangman game complete and ready to play!")
```
